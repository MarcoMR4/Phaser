```python
#Generar csv 


import csv

# Read the input file
with open('input.txt', 'r') as file:
    lines = file.readlines()

# Remove leading and trailing whitespaces, split by space, and create a list of lists
data = [line.strip().split() for line in lines]

# Write the data to a CSV file
with open('output.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerows(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cargar los datos desde el CSV especificando que la primera fila es un encabezado
df = pd.read_csv('datos21.csv', header=None, names=['x1', 'x2', 'target', 'target2'], dtype=float)

# Crear la figura en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos en 3D
scatter = ax.scatter(df['x1'], df['x2'], df['target'], c=df['target'], cmap='viridis', marker='o')

# Etiquetas de los ejes en 3D
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('target')

# Mostrar el gráfico en 3D
ax.view_init(elev=20, azim=30)
plt.show()

```

```python
# Grafica con movimiento 

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import mplcursors

# Cargar los datos desde el CSV especificando que la primera fila es un encabezado
df = pd.read_csv('datos21.csv', header=None, names=['x1', 'x2', 'target', 'target2'], dtype=float)

# Crear la figura en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos en 3D
scatter = ax.scatter(df['x1'], df['x2'], df['target'], c=df['target'], cmap='viridis', marker='o')

# Etiquetas de los ejes en 3D
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('target')

# Estado inicial de la vista
current_elev = 20
current_azim = 30

# Función para manejar eventos de teclado
def on_key(event):
    global current_elev, current_azim
    step = 5  # Ajusta el paso de movimiento
    if event.key == 'up':
        current_elev += step
    elif event.key == 'down':
        current_elev -= step
    elif event.key == 'left':
        current_azim -= step
    elif event.key == 'right':
        current_azim += step
    ax.view_init(elev=current_elev, azim=current_azim)
    fig.canvas.draw()

# Conectar la función on_key al evento key_press_event
fig.canvas.mpl_connect('key_press_event', on_key)

# Habilitar las etiquetas interactivas
mplcursors.cursor(hover=True)

# Mostrar el gráfico en 3D
plt.show()

```

```python
# Grafica 

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importa la proyección 3D

df = pd.read_csv('output.csv', header=None, names=['x1', 'x2', 'y1', 'y2', 'target1', 'target2'], dtype=float)

# Crear la figura en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter1 = ax.scatter(df['x1'], df['x2'], c=df['target1'], cmap='viridis', marker='o', label='Salto')
colorbar1 = plt.colorbar(scatter1)
colorbar1.set_label('Salto')

scatter2 = ax.scatter(df['x1'], df['y2'], df['y1'], c=df['target2'], cmap='plasma', marker='^', label='Derecha')
colorbar2 = plt.colorbar(scatter2)
colorbar2.set_label('Derecha')

# Etiquetas de los ejes en 3D
ax.set_xlabel('distancia')
ax.set_ylabel('velocidad')
ax.set_zlabel('altura')  

# Mostrar el gráfico en 3D
plt.legend()
plt.show()
```
